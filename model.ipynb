```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad_data = pd.read_csv('advertising (1).csv')
```

```python
ad_data.head()
```

```python
ad_data['Ad Topic Line'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
NBmodel = nb.fit(X_train, y_train)
NBPredictions = NBmodel.predict(X_test)
# print(NBPredictions)
acc2 = accuracy_score(y_test, NBPredictions)
print("Improved Accuracy : ",acc2*100)
```

```python
import pickle
with open('NB_train','wb') as m:
    pickle.dump(NBmodel,m)
```
